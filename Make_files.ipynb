```python
import os
import pandas as pd

Data_path = "/content/drive/MyDrive/SIH/Image_Dataset"
Metadata = "/content/drive/MyDrive/SIH/metadata.csv"

meta_df = pd.read_csv(Metadata)
meta2_df = meta_df[["isic_id", "diagnosis"]]
```

```python
meta2_df["path"] = [os.path.join(Data_path,i+".JPG") for i in meta2_df["isic_id"]]
meta2_df
```

```python
# list of unique Diagnosis or n_classes for the classification
diag_list = [i for i in meta2_df["diagnosis"].unique()]
```

```python
# Making a dictionary with destination path
path_dict = {}
count_dict = {}

def make_path(dataset, new_path = "/content/drive/MyDrive/SIH/"):
  count=0
  for i in diag_list:
    path_dict[i] = [new_path+dataset+"/"+i+"/"+j[0]+".JPG" for j in meta2_df.values if j[1] == i]
    count_dict[i] = len(path_dict[i])
    count += len(path_dict[i])
  #print(count)

make_path("Train")
#print(path_dict["actinic keratosis"])
#print(count_dict)
```

```python
# Making all directories and subdirectories
import os
parent_dir_train, parent_dir_test = {}, {}
#os.mkdir("/content/drive/MyDrive/SIH/Train")
#os.mkdir("/content/drive/MyDrive/SIH/Test")
for i in diag_list:
  parent_dir_train[i] = os.path.join("/content/drive/MyDrive/SIH/Train", i)
  parent_dir_test[i] = os.path.join("/content/drive/MyDrive/SIH/Test", i)
  #os.mkdir(os.path.join("/content/drive/MyDrive/SIH/Train", i))
  #os.mkdir(os.path.join("/content/drive/MyDrive/SIH/Test", i))
```

```python
import shutil
# Transferring all files in train directory and subdirectories
def save_files(diag_list, parent_dir_train, path_dict):
  for i in diag_list:
    for j in path_dict[i]:
      old_path = "/content/drive/MyDrive/SIH/Image_Dataset/" + j.split("/")[-1]
      shutil.copyfile(old_path, j)
      #print(old_path, j)


save_files(diag_list, parent_dir_train, path_dict)
```

```python
# Transfer test files from Train directory and subdirectories to Test directory and subdirectories
import shutil
from random import sample

def make_test(path_dict, count_dict):
  test_path_dict = {}
  for i in diag_list:
    test_path_dict[i] = sample(path_dict[i], int(0.2*count_dict[i]))
    #print(i,int(0.2*count_dict[i]))
    for j in test_path_dict[i]:
      test_path = "/content/drive/MyDrive/SIH/Test/"+ "/".join(j.split("/")[6:])
      #print(j, test_path)
      shutil.move(j, test_path)

make_test(path_dict, count_dict)
```
